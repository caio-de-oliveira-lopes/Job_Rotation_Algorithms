```python
import math
import json
import os

#"output-1_hes-c98-BorbaRittModel-NoExtraConstraint-2.json"
path = r'C:\Projetos\Job_Rotation_Algorithms\Outputs'
inputs = r'C:\Projetos\Job_Rotation_Algorithms\Instances\alwabp'

instanceFamilies = ["hes", "ros", "wee", "ton"]
multipliers = [1.05, 1.10, 1.25, 1.5]
models = ["CostaMirallesModel", "BorbaRittModel"]
constraintCases = ["NoExtraConstraint"]
periods = [2]#, "w"]

compiledResults = {}

# family => model => period => multiplier => lista de outputs (casos com resultado) para essas divises

for family in instanceFamilies:
    for model in models: 
        for numberOfPeriods in periods:
            for multiplier in multipliers:
                for i in range(1, 81):
                    instance = f"{str(i)}_{family}"
                    originalCycleTime = cycleTimes[instance]
                    cycleTime = math.floor(originalCycleTime * multiplier)
                    for constraintCase in constraintCases:
                        input_path = os.path.join(inputs, instance)
                        with open(input_path, 'r') as f:
                            file_contents = f.readlines()

                        numberOfWorkers = sum(1 for char in file_contents[1] if char == ' ') + 1
                        numberOfTasks = file_contents[0].replace("\n", " ")

                        if (numberOfPeriods == "w"):
                            numberOfPeriods = numberOfWorkers
                        
                        output = f"output-{instance}-c{cycleTime}-{model}-{constraintCase}-{numberOfPeriods}.json"
                        
                        output_path = os.path.join(path, output);
                        if os.path.exists(output_path):
                            with open(output_path, 'r') as f:
                                data = json.load(f)
                            if (family not in compiledResults):
                                compiledResults[family] = {}
                            if (model not in compiledResults[family]):
                                compiledResults[family][model] = {}
                            if (numberOfPeriods not in compiledResults[family][model]):           
                                compiledResults[family][model][numberOfPeriods] = {}
                            if (multiplier not in compiledResults[family][model][numberOfPeriods]):
                                compiledResults[family][model][numberOfPeriods][multiplier] = {}
                            if ("OF" not in compiledResults[family][model][numberOfPeriods][multiplier]):
                                compiledResults[family][model][numberOfPeriods][multiplier]["OF"] = []
                            if ("time(s)" not in compiledResults[family][model][numberOfPeriods][multiplier]):
                                compiledResults[family][model][numberOfPeriods][multiplier]["time(s)"] = []
                            compiledResults[family][model][numberOfPeriods][multiplier]["OF"].append(data["OF"])
                            compiledResults[family][model][numberOfPeriods][multiplier]["time(s)"].append(data["ExecutionTimeMs"]/1000)

from statistics import mean 

for family in compiledResults.keys():
    for model in compiledResults[family].keys(): 
        for numberOfPeriods in compiledResults[family][model]:
            for multiplier in multipliers:
                try:
                    maxOF = max(compiledResults[family][model][numberOfPeriods][multiplier]["OF"])
                    gaps = [(maxOF - i)/i for i in compiledResults[family][model][numberOfPeriods][multiplier]["OF"]]
                    gap = mean(gaps)
                    
                    averageTime = mean(compiledResults[family][model][numberOfPeriods][multiplier]["time(s)"])
                    maxTime = max(compiledResults[family][model][numberOfPeriods][multiplier]["time(s)"])
                    print(f"{family}-{model}-t({numberOfPeriods})-p({multiplier})")
                    print(f"Gap: {gap}")
                    print(f"t(s): {averageTime}")
                    print(f"t_max(s): {maxTime}")
                    print("-----------------------------------")
                except:
                    print(f"CASE FAILED: {family}-{model}-t({numberOfPeriods})-p({multiplier})")
                    print("-----------------------------------")
```

```python
import os
import matplotlib.pyplot as plt
import gurobi_logtools as glt
import math

# Funo para ler os arquivos e calcular os percentuais
def calcular_percentuais(diretorio):
    # Inicializar contadores
    gaps_costa = []
    gaps_borba = []
    checked = []
    # Percorrer os arquivos no diretrio
    for arquivo in os.listdir(diretorio):
        if arquivo.endswith(".log"):
            if arquivo in checked:
                continue

            checked.append(arquivo)
            if "CostaMirallesModel" in arquivo:
                modelo = "CostaMirallesModel"
            elif "BorbaRittModel" in arquivo:
                modelo = "BorbaRittModel"
            else:
                modelo = None

            # Usar gurobi-logtools para analisar o arquivo de log
            results = glt.parse(os.path.join(diretorio, arquivo))
            summary = results.summary()

            try:
                status = summary['Status'][0]
                gap_modelo = summary['MIPGap'][0] if status != "OPTIMAL" else 0
            except (KeyError, IndexError):
                continue

            if status != "OPTIMAL" and modelo:
                if modelo == "CostaMirallesModel":
                    outro_modelo = "BorbaRittModel"
                elif modelo == "BorbaRittModel":
                    outro_modelo = "CostaMirallesModel"
                
                outro_arquivo = str(arquivo).replace(modelo, outro_modelo)
                checked.append(outro_arquivo)

                # Usar gurobi-logtools para analisar o arquivo de log
                results2 = glt.parse(os.path.join(diretorio, arquivo))
                summary2 = results2.summary()
                try:
                    status = summary2['Status'][0]
                    if status == "OPTIMAL": continue
                    gap_outro_modelo = summary2['MIPGap'][0] if status != "OPTIMAL" else 0
                except (KeyError, IndexError):
                    continue

                if modelo == "CostaMirallesModel":
                    gaps_costa.append(gap_modelo)
                    gaps_borba.append(gap_outro_modelo)
                elif modelo == "BorbaRittModel":
                    gaps_borba.append(gap_modelo)
                    gaps_costa.append(gap_outro_modelo)

    return gaps_costa, gaps_borba

# Funo para agrupar os gaps em grupos de 5%
def agrupar_gaps(gaps):
    grupos = {}
    for gap in gaps:
        if gap is None or math.isnan(gap): continue
        grupo = round(round(gap / 0.05) * 0.05, 2)  # Arredondamento para cima
        grupos[grupo] = grupos.get(grupo, 0) + 1
    return grupos
```

```python
# Diretrio onde os arquivos esto localizados
diretorio = r'C:\Projetos\Job_Rotation_Algorithms\Outputs\GurobiLogs'

# Calcular percentuais
gaps_costa, gaps_borba = calcular_percentuais(diretorio)
```
