```python
import os
import matplotlib.pyplot as plt
import re
import json
import math

dict_instancias = {}
instanceFamilies = ["ros", "hes", "ton", "wee"]
multipliers = [1.05, 1.10, 1.25, 1.5]
models = ["CostaMirallesModel", "BorbaRittModel"]
constraintCases = ["NoExtraConstraint"]
periods = ["2"]
diretorio = r'C:\Projetos\Job_Rotation_Algorithms\Outputs\GurobiLogs'
checked = []

for family in instanceFamilies:
    dict_instancias[family] = {}
    for multiplier in multipliers:
        dict_instancias[family][multiplier] = {
            "#Inst": 0,
            "gaps_M1": [],
            "gaps_M2": [],
            "paths_M1": [],
            "paths_M2": [],
        }

for family in instanceFamilies:
    for arquivo in os.listdir(diretorio):
        if (family not in arquivo): continue
        for multiplier in multipliers:
            if arquivo.endswith(".log"):
                if arquivo+str(multiplier) in checked:
                    continue

                checked.append(arquivo+str(multiplier))
                cycleTime = int(str(arquivo).split("-c")[1].split("-")[0])
                instance = str(arquivo).split("-c")[0].split("-")[-1]
                period = str(arquivo).split("-")[-1].replace(".log", "")                
                recomendedCycleTime = cycleTimes[instance]
                
                if cycleTime != math.floor(recomendedCycleTime * multiplier): continue

                if period not in periods: continue

                try:
                    json_output_path = os.path.join(diretorio.replace("\GurobiLogs", ""), str(arquivo).replace("gurobi_log-", "").replace(".log", ".json"))
                    with open(json_output_path, 'r') as f:
                        multiplier_test = json.load(f)["PercentageAppliedOverOriginalMaximumMeanCycleTime"]
                except Exception as e:
                    continue

                if "CostaMirallesModel" in arquivo:
                    m1 = "CostaMirallesModel"            
                    m2 = "BorbaRittModel"
                elif "BorbaRittModel" in arquivo:
                    m1 = "BorbaRittModel"
                    m2 = "CostaMirallesModel" 
                else:
                    continue

                
                with open(os.path.join(diretorio, arquivo), 'r') as file:
                    # Itere pelas linhas do arquivo
                    for line in file:
                        # Use expressão regular para encontrar o número entre "gap" e "%"
                        match = re.search(r'gap\s+(\d+\.\d+)%', line)
                        if match:
                            # Se houver uma correspondência, imprima o número encontrado
                            gap_m1 = float(match.group(1))

                outro_arquivo = str(arquivo).replace(m1, m2)
                if outro_arquivo+str(multiplier) in checked:
                    continue
                checked.append(outro_arquivo+str(multiplier))

                try:
                    with open(os.path.join(diretorio, outro_arquivo), 'r') as file:
                        # Itere pelas linhas do arquivo
                        for line in file:
                            # Use expressão regular para encontrar o número entre "gap" e "%"
                            match = re.search(r'gap\s+(\d+\.\d+)%', line)
                            if match:
                                # Se houver uma correspondência, imprima o número encontrado
                                gap_m2 = float(match.group(1))
                except:
                    continue

                if (gap_m1 is None or math.isnan(gap_m1) or gap_m1 == "null") or (gap_m2 is None or math.isnan(gap_m2) or gap_m2 == "null"): 
                    continue

                dict_instancias[family][multiplier]["#Inst"] += 1
                if m1 == "CostaMirallesModel":
                    dict_instancias[family][multiplier]["gaps_M1"].append(gap_m1)
                    dict_instancias[family][multiplier]["gaps_M2"].append(gap_m2)
                    dict_instancias[family][multiplier]["paths_M1"].append(os.path.join(diretorio, arquivo))
                    dict_instancias[family][multiplier]["paths_M2"].append(os.path.join(diretorio, outro_arquivo))
                else:
                    dict_instancias[family][multiplier]["gaps_M2"].append(gap_m1)
                    dict_instancias[family][multiplier]["gaps_M1"].append(gap_m2)            
                    dict_instancias[family][multiplier]["paths_M2"].append(os.path.join(diretorio, arquivo))
                    dict_instancias[family][multiplier]["paths_M1"].append(os.path.join(diretorio, outro_arquivo))
        
#print(json.dumps(dict_instancias, indent=4))
```

```python
import matplotlib.pyplot as plt
import gurobi_logtools as glt
import re
import numpy as np

family_order = ['ros', 'hes', 'ton', 'wee']
family_mapping = {
    'hes': 'Heskia',
    'ros': 'Roszieg',
    'wee': 'Wee-Mag',
    'ton': 'Tonge'
}
multiplier_mapping = {
    1.05: '5%',
    1.10: '10%',
    1.25: '25%',
    1.5: '50%'
}

def extract_numbers(file_path):
    explored_nodes = None
    simplex_iterations = None

    with open(file_path, 'r') as file:
        for line in file:
            # Search for "Explored {number} nodes" pattern
            match_explored = re.search(r'Explored (\d+) nodes', line)
            if match_explored:
                explored_nodes = int(match_explored.group(1))

            # Search for "({number} simplex iterations)" pattern
            match_iterations = re.search(r'\((\d+) simplex iterations\)', line)
            if match_iterations:
                simplex_iterations = int(match_iterations.group(1))

            # If both numbers are found, no need to search further
            if explored_nodes is not None and simplex_iterations is not None:
                break

    return [explored_nodes, simplex_iterations]

# Iterate over each family and multiplier combination
for model in ["M1", "M2"]:
    for title in [0, 1]:
        # Initialize an empty list to store the plotted data
        plotted_bars = []
        x_labels = []  # Store X-axis labels
        for family in family_order:
            for multiplier in multipliers:
                try:
                    paths = dict_instancias[family][multiplier][f'paths_{model}']
                    values = []

                    for file in paths:
                        values.append(extract_numbers(file)[title])
                    
                    try:
                        values = np.mean(values) / 10**4
                    except:
                        values = 0
                    
                    # Plot the data
                    bars = plt.bar(len(x_labels), values, label=f"{family_mapping[family]}-{multiplier_mapping[multiplier]}")
                    plotted_bars.append(bars)
                    x_labels.append(f"{family_mapping[family]}-{multiplier_mapping[multiplier]}")  # Store X-axis labels
                except Exception as e:
                    print(f"Error processing {family_mapping[family]}-{multiplier_mapping[multiplier]}: {e}")

        # Customize the plot
        if title == 0:
            plt.ylabel(f'Média de Nós Visitados (x10⁴)')
            plt.title(f'Modelo {model} - Média de Nós Visitados')
        else:
            plt.ylabel('Média de Iterações Simplex (x10⁴)')
            plt.title(f'Modelo {model} - Média de Iterações Simplex')
        
        plt.xlabel('Família-Multiplicador')
        plt.xticks(range(len(x_labels)), x_labels, rotation=90)  # Set X-axis labels
        #plt.yscale('log')  # Set y-axis scale to logarithmic
        plt.legend(bbox_to_anchor=(1, 1))
        # Show the plot
        plt.show()
```
